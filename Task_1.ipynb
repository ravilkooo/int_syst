```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
m = 100
h = 0.5
```

```python
w = 0.2 * np.random.random(3) - 0.1
print(w)
```

```python
xg = np.zeros(m)
yg = np.zeros(m)
w0 = np.zeros((m ,3))
```

```python
for i in range(m):
    p = np.random.random()
    q = np.random.random()
    r = np.random.random()
    if p <= 0.5:
        x = np.array([1, 2*q + 1, 2*r+3])
        d = 1
    else:
        x = np.array([1, 2*q + 4, 2*r+1])
        d = -1
    xg[i] = x[1]
    yg[i] = x[2]
    u = w @ x
    if u >= 0:
        y = 1
    else:
        y = -1
    w = w + h*(d-y) * x
    w0[i] = w
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(range(m), w0[:,0])
ax[0].plot(range(m), w0[:,1])
ax[0].plot(range(m), w0[:,2])
ax[1].scatter(xg, yg)

# Define the coordinates and dimensions of the rectangle (left-bottom x, y, width, height)
rectangle1 = patches.Rectangle((1, 3), 2, 2, linewidth=1, edgecolor='r', facecolor='none')
rectangle2 = patches.Rectangle((4, 1), 2, 2, linewidth=1, edgecolor='b', facecolor='none')
# Add the rectangle to the axis
ax[1].add_patch(rectangle1)
ax[1].add_patch(rectangle2)
x_line = np.linspace(0, 7, 100)
ax[1].plot(x_line, - (w[0] + np.linspace(0, 7, 100)*w[1]) / w[2])

# # Set axis limits
# ax.set_xlim(0, 6)
# ax.set_ylim(0, 6)

# Show the plot
# plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
